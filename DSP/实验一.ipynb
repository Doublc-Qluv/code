```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.signal as signal
```

```python
#解决作图时中文无法显示的问题
import matplotlib.pyplot as plt
plt.rc('font', family='SimHei', size=14)
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问问题
```

```python
A=np.array([1,-0.9])
B=np.array([0.05,0.05])
a=np.ones(8)
x1n=np.hstack((a,np.zeros(50)))
x2n=np.ones(128)
```

```python
system=(B,A,1.0)
t1,h=signal.dimpulse(system)     #脉冲响应
y11n=signal.lfilter(B,A,x1n)       #使用滤波器过滤，标准差分方程的直接形式II转置实现
y12n=signal.lfilter(B,A,x2n)
```

```python
fig,axrr=plt.subplots(2,2,figsize=(14,10))
axrr[0][0].stem(h[0],'C0--')     #C0颜色循环第一种
axrr[0][0].set_title(" 系统单位脉冲响应h(n)",fontsize=18)
axrr[1][0].stem(y11n,'C0--')
axrr[1][0].set_title("系统对R8(n)的响应y1(n)",fontsize=18)#矩形序列
axrr[0][1].stem(y12n,'C0--')
axrr[0][1].set_title("系统对u(n)的响应y2(n)",fontsize=18)#单位阶跃序列
```

```python
############################
x1n=np.ones(8)
h0=np.zeros(10)
h1n=np.hstack((np.ones(10),h0))
h1=np.array([1,2.5,2.5,1])
h2n=np.hstack((h1,np.zeros(10)))
```

```python
#卷积计算
y21n=np.convolve(h1n,x1n)
y22n=np.convolve(h2n,x1n)
```

```python
type(y21n)
```

```python
fig,axrr=plt.subplots(2,2,figsize=(14,10))
axrr[0][0].stem(h1n,'--')
axrr[0][0].set_title("系统单位脉冲响应h1(n)",fontsize=18)
axrr[1][0].stem(y21n,'--')
axrr[1][0].set_title("h1(n)与R8(n)的卷积y21(n)",fontsize=18)
axrr[0][1].stem(h2n,'--')
axrr[0][1].set_title("系统单位脉冲响应h2(n)",fontsize=18)
axrr[1][1].stem(y22n,'--')
axrr[1][1].set_title("h2(n)与R8(n)的卷积y21(n)",fontsize=18)
```

```python
#######################################
un=np.ones(256)
n=np.arange(0,255)
xsin=np.sin(0.014*n)+np.sin(0.4*n)
A=[1,-1.8237,0.9801];
B=[1/100.49,0,-1/100.49]; 
#系统差分方程系数向量B和A
y31n=signal.lfilter(B,A,un)
y32n=signal.lfilter(B,A,xsin)
```

```python
fig,axrr=plt.subplots(2,1,figsize=(14,12))
axrr[0].stem(y31n,'C0--')
axrr[0].set_title("谐振器对u(n)的响应y31(n)",fontsize=18)
axrr[1].stem(y32n,'C0--')
axrr[1].set_title("谐振器对正弦信号的响应y32(n)",fontsize=18)
#sns.despine
```
