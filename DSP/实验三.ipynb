```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#解决作图时中文无法显示的问题
import matplotlib.pyplot as plt
plt.rc('font', family='SimHei', size=14)
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问问题
```

```python
#产生序列向量x1(n)=R4(n)
x1n=np.ones(4)
M=8
xa=np.arange(1,M/2+1)
xb=np.arange(M/2,0,-1)
#产生长度为8的三角波序列x2(n)
x2n=np.hstack((xa,xb))
x3n=np.hstack((xb,xa))
#自变量
f1=np.arange(0,16)/8
f2=np.arange(0,8)/4
```

```python
#计算x1n的8点DFT
X1k8=np.fft.fft(x1n,8);       
#计算x1n的16点DFT
X1k16=np.fft.fft(x1n,16); 
#计算x1n的8点DFT
X2k8=np.fft.fft(x2n,8);     
#计算x1n的16点DFT
X2k16=np.fft.fft(x2n,16);   
#计算x1n的8点DFT
X3k8=np.fft.fft(x3n,8);
#计算x1n的16点
X3k16=np.fft.fft(x3n,16);   
```

```python
#以下绘制幅频特性曲线
fig,axxr=plt.subplots(2,1,figsize=(14,13))
#绘制8点DFT的幅频特性图
axxr[0].stem(f2,X1k8)
axxr[0].set_title("(1a) 8点DFT[x_1(n)]")    #变换区间N=8时，进行频谱分析时得到的幅频特性曲线
axxr[0].set_xlabel("ω/π")
axxr[0].set_ylabel('幅度')
#绘制16点DFT的幅频特性图
axxr[1].stem(f1,X1k16)
axxr[1].set_title('(1b)16点DFT[x_1(n)]')
axxr[1].set_xlabel('ω/π')
axxr[1].set_ylabel('幅度')
```

```python
fig,axxr=plt.subplots(2,2,figsize=(16,12))
#绘制8点DFT的幅频特性图
axxr[0][0].stem(f2,abs(X2k8))
axxr[0][0].set_title("(2a) 8点DFT[x_2(n)]")
axxr[0][0].set_xlabel("ω/π")
axxr[0][0].set_ylabel('幅度')
#绘制16点DFT的幅频特性图
axxr[0][1].stem(f1,abs(X2k16))
axxr[0][1].set_title("(2b)16点DFT[x_2(n)]")
axxr[0][1].set_xlabel("ω/π")
axxr[0][1].set_ylabel('幅度')
#绘制8点DFT的幅频特性图
axxr[1][0].stem(f2,abs(X3k8))
axxr[1][0].set_title("(3a) 8点DFT[x_3(n)]")
axxr[1][0].set_xlabel("ω/π")
axxr[1][0].set_ylabel('幅度')
#绘制16点DFT的幅频特性图
axxr[1][1].stem(f1,abs(X3k16))
axxr[1][1].set_title("(3b)16点DFT[x_3(n)]")
axxr[1][1].set_xlabel("ω/π")
axxr[1][1].set_ylabel('幅度')
```

```python
#实验内容(2) 周期序列谱分析===================
```

```python
#FFT的变换区间N=8
N=8
n=np.arange(N)
x4n=np.cos(np.pi*n/4)
x5n=np.cos(np.pi*n/4)+np.cos(np.pi*n/8);
#计算x4n的8点DFT
X4k8=np.fft.fft(x4n);      
#计算x5n的8点DFT
X5k8=np.fft.fft(x5n);       
#FFT的变换区间N=16
N=16
n=np.arange(N)     
x4n=np.cos(np.pi*n/4);
x5n=np.cos(np.pi*n/4)+np.cos(np.pi*n/8);
#计算x4n的16点DFT
X4k16=np.fft.fft(x4n);      
#计算x5n的16点DFT
X5k16=np.fft.fft(x5n);      
```

```python
fig,axxr=plt.subplots(2,2,figsize=(16,10))
#绘制8点DFT的幅频特性图
axxr[0][0].stem(f2,abs(X4k8))
axxr[0][0].set_title("(4a) 8点DFT[x_4(n)]")
axxr[0][0].set_xlabel("ω/π")
axxr[0][0].set_ylabel('幅度')
#绘制16点DFT的幅频特性图
axxr[0][1].stem(f1,abs(X4k16))
axxr[0][1].set_title("(4b)16点DFT[x_4(n)]")
axxr[0][1].set_xlabel("ω/π")
axxr[0][1].set_ylabel('幅度')
#绘制8点DFT的幅频特性图
axxr[1][0].stem(f2,abs(X5k8))
axxr[1][0].set_title("(5a) 8点DFT[x_5(n)]")
axxr[1][0].set_xlabel("ω/π")
axxr[1][0].set_ylabel('幅度')
#绘制16点DFT的幅频特性图
axxr[1][1].stem(f1,abs(X5k16))
axxr[1][1].set_title("(5b)16点DFT[x_5(n)]")
axxr[1][1].set_xlabel("ω/π")
axxr[1][1].set_ylabel('幅度')
```

```python
#实验内容(3) 模拟周期信号谱分析=================
Fs=64
T=1/Fs;
#FFT的变换区间N=16
N=16
n=np.arange(N)        
x6nT=np.cos(8*np.pi*n*T)+np.cos(16*np.pi*n*T)+np.cos(20*np.pi*n*T);
#对x6(t)16点采样
#计算x6nT的16点DFT
X6k16=np.fft.fft(x6nT); 
#将零频率移到频谱中心 
X6k16=np.fft.fftshift(X6k16); 
##频率分辨率F
Tp=N*T
F=1/Tp; 
#产生16点DFT对应的采样点频率（以零频率为中心）
k=np.arange(-N/2,N/2)
fk=k*F
```

```python
fig,axxr=plt.subplots(1,1,figsize=(12,4))
#绘制16点DFT的幅频特性图
axxr.stem(fk,abs(X6k16),'--')
axxr.set_title("(6a) 16点|DFT[x_6(nT)]|")
axxr.set_xlabel("f(Hz)")
axxr.set_ylabel("幅度")
```
