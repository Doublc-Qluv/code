```python
# coding: utf-8

from tkinter import *

import tkinter.simpledialog as dl
import tkinter.messagebox as mb
```

```python
# setting GUI
root = Tk()
w = Label(root, text = "Guess Number Game")
w.pack()
```

```python
# Welcome Message
mb.showinfo("Welcome", "Welcome Message")

number = 59

while True:
    guess = dl.askinteger("Number", "Enter a number")
    
    if guess == number:
        output = 'Bingo! You guessed it right, but you do not win any prize! haha!'
        mb.showinfo("Hint: ", output)
        break
    elif guess < number:
        output = 'No, the number is a higer than that'
        mb.showinfo("Hint: ", output)
    else:
        output = 'No, the number is a lower than that'
        mb.showinfo("Hint: ",output)
print('Done')
```
